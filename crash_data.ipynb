```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import geopandas as gpd
import os
from datetime import datetime

# show graphs & charts inline
%matplotlib inline
plt.rcParams["figure.figsize"] = (10, 5) # define plot size
saved_style_state = matplotlib.rcParams.copy() # give us a style state to go back to
```

```python
df_crashData = pd.read_csv("data/NSW_Road_Crash_Data_2016-2020.csv")
print("done")
```

```python
df_crashData.columns
```

```python
df_crashData["Degree of crash - detailed"].value_counts().plot(kind="bar")

plt.title("Types of crashes", fontsize=20)
plt.xlabel("Crash Types")
plt.ylabel("Number of crashes")
```

```python
df_crashData["Month of crash"].value_counts().plot(kind="bar")

plt.title("Crash data by month", fontsize=20)
plt.xlabel("Months")
plt.ylabel("Number of crashes")
```

```python
df_crashData["Day of week of crash"].value_counts().plot(kind="bar")

plt.title("Crash data by day of week", fontsize=20)
plt.xlabel("Days")
plt.ylabel("Number of crashes")
```

```python
df_crashData["Time of crash - Two-hour intervals"].value_counts().plot(kind="bar")

plt.title("Crash data by month", fontsize=20)
plt.xlabel("Crash times - Two-hour intervals")
plt.ylabel("Number of crashes")
```

```python
df_crashData["Surface condition"].value_counts().plot(kind="bar")

plt.title("Crashes by surface condition", fontsize=20)
plt.xlabel("Surface condition")
plt.ylabel("Number of crashes")
```
